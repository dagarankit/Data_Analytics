```python
import seaborn as sns
%matplotlib inline
```

```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tips.info()
```

```python
type(tips)
```

```python
sns.distplot(tips['total_bill'],kde=True) # kde  kernal density estimator univariate 
```

```python
# bivariate  
#corresponding x and y plots graph aside
sns.jointplot(x='total_bill',y='tip',data=tips,kind='scatter')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips,hue='sex',palette='rainbow')
```

```python
sns.rugplot(tips['total_bill']) # draws dash mark for every point "Univariate distribution" matplotlib.savepic
```

```python
sns.countplot(x='sex',data=tips)
```

```python
sns.countplot(x='smoker',data=tips)
```

```python
sns.boxplot(data=tips,palette='rainbow',orient='h') # u not need to pass the x and y for orient 
```

```python
sns.boxplot(x='day',y='total_bill',hue='smoker',data=tips,palette='rainbow')
```

```python
sns.violinplot(data=tips,palette='rainbow',orient='h') ## much broader denser it will be 
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,palette='rainbow') ## much broader denser it will be 
```

```python
sns.violinplot(x='total_bill',y='sex',data=tips,palette='rainbow') ## much broader denser it will be 
```
