```python
#to plot death count wrt date
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style('darkgrid')
ls=covid['Date'].value_counts()
ls=ls[0:10]
plt.figure(figsize=(20,6))
death_count=sns.barplot(ls.index,ls.values)
plt.title('Death count wrt date')
plt.ylabel('Count',fontsize=12)
plt.xlabel('Date',fontsize=15)
death_count.set_xticklabels(rotation=30,labels=ls.index,fontsize=10)
```

```python
# death_count wrt Country/Region
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style('darkgrid')
ls=covid['Country/Region'].value_counts()
ls=ls[50:100]
plt.figure(figsize=(20,6))
death_count=sns.barplot(ls.index,ls.values)
plt.title('Death count wrt date')
plt.ylabel('Count',fontsize=12)
plt.xlabel('Country/Region',fontsize=15)
death_count.set_xticklabels(rotation=30,labels=ls.index,fontsize=10)
```

```python

import matplotlib.pyplot as plt
print("count of restaurents at unique locations")
locationCount=covid['location'].value_counts().sort_values(ascending=True)
locationCount

fig=plt.figure(figsize=(20,40))
locationCount.plot(kind="barh",fontsize=20) # barh for representing horizontal bar
plt.ylabel("Location names",fontsize=50,color="red",fontweight='bold')
plt.title("LOCATION VS RESTAURANT COUNT GRAPH",fontsize=40,color="BLACK")
plt.show()
```
