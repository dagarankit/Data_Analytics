```python
values=[12,55,4,32,14]
colors = ['r','g','b','c','m']
explode = [0.2,0,0.5,0,0]
labels=['India','United States','Russia','China','Europe']
plt.pie(values,colors=colors, labels=labels,explode=explode,shadow=True,autopct='%1.3f%%',counterclock=True)
plt.title('Students Locations')
plt.show()
```
